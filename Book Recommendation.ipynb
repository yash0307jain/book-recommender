```python
x = ratings['userId'].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
ratings = ratings[ratings['userId'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head(2)
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head(2)
```

```python
number_rating.rename(columns={'rating': 'number of rating'}, inplace=True)
```

```python
number_rating.head(2)
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['number of rating'] >= 50]
```

```python
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['userId', 'title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='userId', index='title', values='rating')
```

```python
book_pivot.shape
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237, :].values.reshape(1, -1)
```

```python
distance, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1))
```

```python
distance
```

```python
suggestions
```

```python
book_pivot.index[7]
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
movies_dic = {}
for i in range(len(book_pivot.index)):
    movies_dic[book_pivot.index[i]] = i
```

```python
movies_dic
```

```python
def recommender(movie):
    movie_id = movies_dic[movie]
    distance, suggestions = model.kneighbors(book_pivot.iloc[movie_id, :].values.reshape(1, -1))
    for i in range(len(suggestions)):
        print(book_pivot.index[suggestions[i]])
```

```python
recommender('Harry Potter and the Chamber of Secrets (Book 2)')
```
